```python
import face_recognition
import cv2
from google.colab.patches import cv2_imshow
```

```python
img=face_recognition.load_image_file('/content/Valentino_Rossi_2010_Qatar.jpg')
VR_encoding=face_recognition.face_encodings(img)[0]
known_encoding=[VR_encoding]
known_names=['VR']
```

```python
known_encoding
```

```python
img1=face_recognition.load_image_file('/content/Valentino_Rossi_2017.jpg')
img_enc1=face_recognition.face_encodings(img)[0]
print(img_enc1)


```

```python
results=face_recognition.compare_faces(known_encoding,img_enc1)
results
```

```python
for i in range(len(results)):
  if results[i]:
    name=known_names[i]
    (top,right,bottom,left)=face_recognition.face_locations(img1)[0]
    cv2.rectangle(img1,(left,top),(right,bottom),(0,0,255),2)
    cv2.putText(img1,name,(left+2,bottom+20),cv2.FONT_HERSHEY_PLAIN,1,(0,255,255),1)
    cv2_imshow(img1)
```
